# 1: Regresi Linear

Misalkan $A \times x = y$ adalah model regresi, dimana $x$ adalah vektor untuk nilai ```intercept``` dan ```slope```. Maka solusi dari persamaan regresi tersebut adalah:

$$x = (A^{T}A)^{-1}A^{T}y$$

dimana $A$ adalah matriks model dan $y$ adalah vektor response. 

1. Buatlah model regresi diatas hanya dengan menggunakan menggunakan metode inverse matriks diatas
2. Bandingkan hasilnya dengan menggunakan ```method``` ```linalg.lstsq``` dari NumPy
3. Bandingkan hasilnya dengan menggunakan ```stats.linregress``` dari SciPy
4. Plot hasil regressinya

````python
# Membuat fungsi untuk model regresi dengan metode inverse matrix

def regresi_inverse_matriks(var_x, var_y):
    import numpy as np
    from numpy.linalg import inv 
    
    length_data = len(x)
    
    one = np.full(length_data, 1)
    
    A = np.column_stack((one, var_x))
    
    A_transpose = A.transpose()
    
    #computation
    output = inv(A_transpose @ A) @ A_transpose @ var_y
    
    print(f'intercept = {round(output[0],3)} \nslope = {round(output[1],3)}')
    
    return
````

````python
# Metode linalg.lstsq
from numpy.linalg import lstsq
import numpy as np

A = np.column_stack((np.ones(len(variabel_x)), variabel_x))

intercept, slope = lstsq(A, variabel_y, rcond=-1)[0]
print(intercept, slope)
````

````python
# menggunakan stats.linregress dari SciPy
from scipy.stats import linregress

slope, intercept, rvalue, pvalue, stderror = linregress(variabel_x, variabel_y)

print(intercept, slope)
````
